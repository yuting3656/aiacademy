```python
#sqrt : 對raw.data做開平方根再乘十的轉換，並存在sqrt10.data中。
sqrt10data = np.sqrt(rawdata) * 10

#plot(raw.data, sqrt10.data, main = "sqrt10", type = "l", lwd = 2, col = "blue", asp = 1)
#abline(a = 0, b = 1)
plt.scatter(rawdata, sqrt10data)
```

```python
#/max : 將raw.data除以其中最大值，讓所有數值介於0-1之間，並存在trans.max.data中。
maxdata = max(rawdata)
transmaxdata = rawdata/max(rawdata)

#plot(raw.data, trans.max.data, main = "/max", type = "l", lwd = 2, col = "blue")
plt.scatter(rawdata, transmaxdata)
```

```python
#/sum : 將raw.data除以全部總和，讓所有數值轉換成百分比，並存在trans.sum.data中。
sumdata = sum(rawdata)
transsumdata = rawdata/sum(rawdata)  #Species profile transformation

#plot(raw.data, trans.sum.data, main = "/sum", type = "l", lwd = 2, col = "blue")
plt.scatter(rawdata, transsumdata)
```

```python
#Chord transformed : 將資料除上平方和的開根號。
transchorddata = rawdata/np.sqrt(sum(rawdata**2)) #length of 1, Chord transformation

#plot(raw.data, trans.chord.data, main = "norm (Chord)", type = "l", lwd = 2, col = "blue")
plt.scatter(rawdata, transchorddata)
```

```python
#Hellinger transformed : 將資料除以全部總和，再開根號。
transhellingerdata = np.sqrt(rawdata/sum(rawdata)) #Hellinger transformation

#plot(raw.data, trans.hellinger.data, main = "Hellinger", type = "l", lwd = 2, col = "blue")
plt.scatter(rawdata, transhellingerdata)
```
