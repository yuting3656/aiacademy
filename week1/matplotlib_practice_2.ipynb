### Using axes

```python
fig = plt.figure()

# new Figure axes (left, bottom, width, heigh) (0~1)
axes1 = fig.add_axes([0.05, 0.05, 0.425, 0.8])
axes2 = fig.add_axes([0.55, 0.05, 0.425, 0.8])

# axes1
axes1.plot(edu.Year, edu["Physical Sciences"], color="blue")
axes1.set_xlabel('Year')
axes1.set_ylabel('Physical Sciences')
axes1.set_title('Year vs. PS')

# axes2
axes2.plot(edu.Year, edu["Computer Science"], color="red")
axes2.set_xlabel('Year')
axes2.set_ylabel('CS')
axes2.set_title('Year vs. CS')

#  
fig.suptitle('Comparision', fontsize=16, fontweight="bold")
plt.show()
```

```python
import pandas as pd
import numpy as np
import math
import decimal
import matplotlib.pyplot as plt
data = pd.read_excel("matplotlib_csv_file/AIA_City.xlsx")


## 


plt.style.use("fivethirtyeight")

x = data[""]
x.unique()
label_x = ['Taipei', 'New Taipei City', 'Kaohsiung', 'Taoyuan', 'Taichung', 'Hsinchu', 'Keelung', 'Tainan', 'Pingtung',
           'Chiayi','Yunlin', 'Nantou', 'Miaoli']
#'', '', '', '', '','', '', '', '', '', '', '', ''
label_x = label_x[::-1]
print(label_x)

#percentage
count_x = x.value_counts().values
percent_x = count_x/sum(count_x)
percent_x = map(lambda x: round((round(x,5))*100,1), percent_x)
percent_x = list(percent_x)


#round
y = []
for i in range(len(percent_x)):
    number = decimal.Decimal(str(percent_x[i]))
    y.append(float(number))
y = y[::-1]




fig, ax = plt.subplots()    
width = 0.7 # the width of the bars 
num = np.arange(len(y))  # the x locations for the groups
my_color = ['#F67770','#DF8921','#BD9B21','#8DAA20','#2EB51F','#1EBEA9','#1DAEF9','#1FBBD8','#F767DB','#D47AFB',
            '#8C96FC','#FD68AC','#F67770']

my_color = my_color[::-1]
ax.barh(num, y, width, color = my_color)
ax.set_yticks(num+0.1)
ax.set_yticklabels(label_x, minor=False)
for i, v in enumerate(y):
    ax.text(v + 0.25, i-0.2, str(v)+ "%", color='blue', fontweight='bold', fontsize = 25)

    
# plt.title('residence',size = 40)
plt.xlabel('Percentage', size = 38)
plt.tick_params(axis='both', labelsize=30)
plt.xlim(0,40)
plt.tight_layout()
fig.set_size_inches(25, 12)
fig.savefig('tech_hometown.png', dpi=400)


plt.show()
```
